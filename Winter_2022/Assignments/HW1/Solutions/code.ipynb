**span of matrix A :**

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize= (8,6))

ax1 = fig.add_subplot(projection='3d')

for i in np.arange(-5, 5):
  for j in np.arange(-5, 5):
    for k in np.arange(-5, 5):
      ax1.scatter(1*i + 1*j + 1*k, 2*i + 2*j + 2*k, 1*i + 1*j + 1*k, c='#2B67FF')

ax1.set_xlabel('x', fontsize=18)
ax1.set_ylabel('y', fontsize=18)
ax1.set_zlabel('z', fontsize=18);
```

**span of matrix B :**

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize= (8,6))

ax1 = fig.add_subplot(projection='3d')

for i in np.arange(-5, 5):
  for j in np.arange(-5, 5):
    for k in np.arange(-5, 5):
      ax1.scatter(1*i + 1*j + 1*k, 2*i - 2*j - 2*k, 1*i - 1*j - 1*k, c='#2B67FF')

ax1.set_xlabel('x', fontsize=18)
ax1.set_ylabel('y', fontsize=18)
ax1.set_zlabel('z', fontsize=18);
```

اسپن سه نمودار، تشکیل یک صفحه را داده است که با پاسخ ارائه شده در بخش اول تطابق دارد

**span of matrix C :**

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize= (8,6))

ax1 = fig.add_subplot(projection='3d')

for i in np.arange(-3, 3):
  for j in np.arange(-3, 3):
    for k in np.arange(-3, 3):
      ax1.scatter(1*i + 1*j - 1*k, 2*i - 2*j + 2*k, 1*i + 1*j + 1*k, c='#2B67FF')

ax1.view_init(azim=20, elev=30)

ax1.set_xlabel('x', fontsize=18)
ax1.set_ylabel('y', fontsize=18)
ax1.set_zlabel('z', fontsize=18);
```
