## Part 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[1, -1],
              [2, 2]])
```

```python
origin = np.array([[0, 0],
                   [0, 0]])
```

```python
x_min, x_max = -5, 5
y_min, y_max = -5, 5

scale = 1
```

```python
def draw_grid(x_min, x_max, y_min, y_max):
    
    fig = plt.figure()

    # 1Ã—1 grid, first subplot
    ax = fig.add_subplot(1, 1, 1) 

    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)

    grid_x_ticks = np.arange(x_min, x_max + 1, 1)
    grid_y_ticks = np.arange(y_min, y_max + 1, 1)

    ax.set_xticks(grid_x_ticks)
    ax.set_yticks(grid_y_ticks)

    ax.grid()
    
    return ax
```

```python
ax = draw_grid(x_min, x_max, y_min, y_max)

plt.show()
```

```python
ax = draw_grid(x_min, x_max, y_min, y_max)

ax.quiver(*origin, *A, scale=scale, scale_units='xy', angles='xy', color=['g', 'r'])

plt.show()
```

```python
ax = draw_grid(x_min, x_max, y_min, y_max)

scaler = np.arange(-5., 5., .001)
scaler = scaler.reshape(scaler.shape[0], 1)

np.random.shuffle(scaler)
V1 = scaler*A[:, 0]

np.random.shuffle(scaler)
V2 = scaler*A[:, 1]

span = V1 + V2

ax.plot(span[:, 0], span[:, 1], 'b.')
    
ax.quiver(*origin, *A, scale=scale, scale_units='xy', angles='xy', color=['g', 'r'])

plt.show()
```

### Linear Combination

```python
ax = draw_grid(x_min, x_max, y_min, y_max)

scaler = np.arange(-10., 10., .1)
scaler = scaler.reshape(scaler.shape[0], 1)

added_vectors = A[:, 0] + scaler*A[:, 1]

ax.plot(added_vectors[:, 0], added_vectors[:, 1], 'b-')
    
ax.quiver(*origin, *A, scale=scale, scale_units='xy', angles='xy', color=['g', 'r'])

plt.show()
```
